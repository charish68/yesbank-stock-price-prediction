#### 2. Perform an appropriate statistical test.

````python
import pandas as pd
from scipy.stats import ttest_ind

# Ensure 'Date' column is datetime
df['Date'] = pd.to_datetime(df['Date'], format='%b-%y')

# Calculate Volatility
df['Volatility'] = df['High'] - df['Low']

# Split dataset
vol_before = df[df['Date'] < '2020-03-01']['Volatility']
vol_after = df[df['Date'] >= '2020-03-01']['Volatility']

# Perform Independent Two-Sample t-test (Welch‚Äôs t-test)
t_stat, p_value_two_tailed = ttest_ind(vol_before, vol_after, equal_var=False)

# Convert to one-tailed test by dividing the p-value by 2
p_value_one_tailed = p_value_two_tailed / 2

# Print results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value (one-tailed): {p_value_one_tailed:.4f}")
````

##### Which statistical test have you done to obtain P-Value?

We performed an Independent Two-Sample t-test (Welch‚Äôs t-test) to obtain the P-value.

Used scipy.stats.ttest_ind()

Compared average volatility before and after March 2020

Used one-tailed version by dividing the default two-tailed p-value by 2

üîπ 3. Unequal Sample Sizes & Variances:

### 2. Handling Outliers

````python
import pandas as pd
from scipy.stats import zscore
import matplotlib.pyplot as plt
import seaborn as sns

# Sample assumption: your dataset is already loaded as `df`
# If not, load it using: df = pd.read_csv("your_file.csv")

# Step 1: IQR-Based Outlier Detection
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Detect outliers in 'Close' column
outliers_close = detect_outliers_iqr(df, 'Close')
print("Number of outliers in 'Close':", len(outliers_close))


# Step 2: Outlier Capping (Winsorization)
def cap_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Cap the values using clip
    data[column] = data[column].clip(lower=lower_bound, upper=upper_bound)
    return data

# Apply capping on relevant columns
columns_to_cap = ['Open', 'High', 'Low', 'Close']
for col in columns_to_cap:
    df = cap_outliers(df, col)


# Step 3: Optional - Remove Outliers Using Z-score
# Uncomment below lines if you prefer to drop outliers
# def remove_outliers_zscore(data, column):
#     z_scores = zscore(data[column])
#     abs_z = abs(z_scores)
#     return data[abs_z < 3]

# df = remove_outliers_zscore(df, 'Close')


# Step 4: Visualize Boxplots After Outlier Treatment
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[columns_to_cap])
plt.title("Boxplot After Outlier Capping")
plt.xticks(rotation=45)
plt.show()
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

In this project, we addressed the issue of outliers using two key techniques: IQR-based capping (Winsorization) and Z-score filtering. The primary method applied was IQR capping, where values beyond 1.5 times the interquartile range from the first and third quartiles were identified as outliers. These extreme values were then capped at the lower and upper bounds rather than being removed. This technique was selected because it retains all rows, preserving the continuity and temporal structure of the time-series stock data, which is critical for financial forecasting. In addition, Z-score filtering was explored as an optional method to detect and remove extreme values by checking whether they fall beyond three standard deviations from the mean. However, we opted for capping as the preferred approach, as sudden spikes or drops in stock prices are often genuine reflections of market conditions. Capping helped reduce the skewness caused by such outliers without discarding meaningful data points, ensuring the model remains robust yet realistic.

Risk managers reduce exposure to unpredictable swings.

Finance teams forecast revenue from holdings with more precision.

What It Measures:
MSE calculates the average squared difference between actual and predicted values.

Interpretation in Business Terms:

Punishes larger errors more heavily than MAE.

Helps identify months where predictions were significantly off.

Business Impact:
Lower MSE indicates your model avoids big financial forecasting mistakes, which can:

Prevent huge investment losses.

Help CFOs and analysts improve forecasting systems.

Build stakeholder trust in AI-assisted financial tools.

üîπ 3. Root Mean Squared Error (RMSE)

RMSE is the square root of MSE, giving an error metric in the original units.

Interpretation in Business Terms:

Like MAE, but gives more weight to large outliers.

Used when minimizing large deviations is critical.

Business Impact:
Lower RMSE means greater confidence in the model during volatile periods, helping:

Portfolio managers adjust quickly during market crashes.

Institutions set more accurate price thresholds for automation or alerts.

üîπ 4. R-squared Score (R¬≤)
What It Measures:
R¬≤ represents the percentage of variance in the target variable (Close price) explained by the model.

Interpretation in Business Terms:

A value closer to 1.0 means the model is doing a great job at capturing patterns.

Lower values mean your features are missing key drivers.

Business Impact:
A higher R¬≤ builds trust in ML adoption:

Financial analysts can rely more on automated tools.

Helps identify which features (Open, High, Low) are strong predictors.

Supports internal reporting and regulatory compliance
